```python
import random


N=100
mu=0
sig=1

x=0
b=[]

for i in range (N):
    b.append(x)
    w=random.gauss(mu, sig)
    x=x+w

w=100 #window length

ma=[]

for i in range(len(b)):
    s=0 # sum of numbers i have counted so far is zero initially
    n=0 #count is zero initially
    for j in range(w):
        if i-j<0: #önceki sayılara bakıyor bu şekilde
            break;
        s=s+b[i-j]
        n=n+1
    ma.append(s/n)
    
plt.plot(b)
plt.plot(ma)
plt.show()
```
